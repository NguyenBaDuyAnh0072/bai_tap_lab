```python
import time
import threading

class SimpleTask:
    def __init__(self, counter):
        self.counter = counter

    def run_task(self):
        for _ in range(16):
            time.sleep(2)
            self.counter += 1
            print(f"Counter has increased to: {self.counter}")

def main():
    counter = 0
    tasks = [threading.Thread(target=SimpleTask(counter).run_task) for _ in range(4)]
    for task in tasks:
        task.start()
    for task in tasks:
        task.join()

if __name__ == "__main__":
    main()
```
